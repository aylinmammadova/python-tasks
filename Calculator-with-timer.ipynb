```python
import random
import time
import threading

def generate_expression():
    MIN_NUM, MAX_NUM = 0, 20
    OPS = ['+', '-', '*', '/']
    
    left = random.randint(MIN_NUM,MAX_NUM)
    operator = random.choice(OPS)
    right = random.randint(MIN_NUM,MAX_NUM)
    if right == 0:
        right = random.randint(MIN_NUM,MAX_NUM)
    else:
        return '{0!s}{1}{2!s}'.format(left, operator, right)

def evaluate():
    expression = generate_expression()
    if expression == None:
        expression = generate_expression()
    
    print('Predict the value of the following expression: ', expression)
    expression = expression.split(' ')[-1]
    return eval(expression)

def countdown():
    global STOP_TIMER, TIME
    STOP_TIMER = False
    TIME = 5
    while not STOP_TIMER:
        print('Time left: {0}\r'.format(TIME), end='')
        time.sleep(1)
        TIME -= 1
        if STOP_TIMER or TIME<0:
            break

end_game = False
score = 0

while not end_game :
    countdown_thread = threading.Thread(target=countdown)
    question = evaluate()
    countdown_thread.start()
    answer = input('Enter your answer: ')
    
    try:
        answer = int(answer)
    except:
        answer = float(answer)
    
    if TIME<0:
        print("Time is over.")
        end_game = True
    
    if answer and (TIME>=0):
        STOP_TIMER = True
        countdown_thread.join()
            
        if question == answer:
            score += 1
        else:
            end_game = True
        
    if end_game == True:
        STOP_TIMER = True
        countdown_thread.join()
        
        print(f"The correct answer was: {question}")
        print(f"Final score is: {score}")
```
